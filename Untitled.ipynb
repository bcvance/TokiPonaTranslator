```python
def line_to_key_value(line):
    line = line.strip()
    if '(' in line:
        line = line[0:line.find("(")] + line[line.find(")")+1:]
    key = line.split()[0]
    values = ' '.join(line.split()[1:]).replace(';',',').replace(' or',',').split(',')

    return key, values
    
    

line_to_key_value('PARTICLE (emphasis, emotion or confirmation)')   
    
```

```python
tokipona_words = [] 
for line in dictionary_lines: 
    if line[0] == " ": #
        key,value = line_to_key_value(line)
        for word in tokipona_words:
            if word in dictionary:
                dictionary[word][key] = value
            else:
                dictionary[word] = {key:value} 
    else: 
        if "or" in line:
            tokipona_words = [word.strip() for word in line.split("or")]
        else:
            tokipona_words = [line]
print(dictionary)
        
        
        
```

```python
# input sentence -> phrase -> translated sentence
#'ijo li moku e mi'

# class Phrase:
#     def __init__(self):
#         self.subject = None
#         self.verb = None
#         self.verb_or_adjective = None
#         self.object = None
#         self.context = None

# def tokenize_and_tag(line):
"""
split on 'la' -> before 0 context
split rest on 'li' - > 0 becomes subject OR if first word is mi or sina
split rest on 'e' 
   -> 
   
'wawa': {'ADJECTIVE': ['strong', ' powerful', ' confident', ' sure', ' energetic', ' intense']}, 
'weka': {'ADJECTIVE': ['absent', ' away', ' ignored']}, 
'wile': {'PRE-VERB': ['must', ' need', ' require', ' should', ' want', ' wish']}}

"""
```

```python
import random

input_lines = poem.replace(",", ".").split(".")
output = [] # an ala ken sona pilin ike
for line in input_lines: # line = jan ala li ken sona e pilin ike mi. i = 8
    #     phrase = tokenize_and_tag(line)
    #     translated_sentence = translate(phrase)
    # translation = []
    choice = None
    last_particle = None
    last_particle_index = None # 5
    tokens = line.split()
    for i, token in enumerate(tokens): 
        #print(output)
        if token in ["la", "li", "e"]:
            #print(f"hello we are here{token}")
            last_particle = token
            last_particle_index = i
            choice = None
        else:
            #print(f"HERE {token}")
            if token in dictionary:
                #print(f"hi {token}")
                if choice and choice == "PRE-VERB": # pre verb followed by verb rule
                    if "VERB" in dictionary[token]:
                        choice = "VERB"
                        word = random.choice(dictionary[token][choice])
                        #print(f"word: {word}")
                        output.append(word)
                    else:
                        #print(f"VERB not in dictionary[{token}]")
                        choice = random.choice(list(dictionary[token]))
                        word = random.choice(dictionary[token][choice])
                        #print(f"word: {word}")
                        output.append(word)
                elif "SUBJECT-PRONOUN" in dictionary[token]:
                    if last_particle and last_particle == "e" and i > 0 and not tokens[i-1] == "e": # possessive adj rule
                        choice = "POSSESSIVE-PRONOUN"
                        word = random.choice(dictionary[token][choice])
                        # insert word
                        number_of_preceeding_nouns = i - last_particle_index - 1
                        output = output[:-number_of_preceeding_nouns] + [word] + output[-number_of_preceeding_nouns:]
                        
                    elif i > 0 and tokens[i-1] == "e": # object pronoun rule
                        choice = "OBJECT-PRONOUN"
                        word = random.choice(dictionary[token][choice])
                        #print(f"word: {word}")
                        output.append(word)
                    else: 
                        choice = "SUBJECT-PRONOUN"
                        word = random.choice(dictionary[token][choice])
                        #print(f"word: {word}")
                        output.append(word)
                    
                else:
                    choice = random.choice(list(dictionary[token]))
                    word = random.choice(dictionary[token][choice])
                    #print(f"word: {word}")
                    output.append(word)
#                 print(token, dictionary[token])
    output.append("\n")
#     print("\n")
output = ' '.join(output)
print(output)
print(poem)
                
                
    
```
